### Thresholding

Thresholding is a simple, yet effective, way of partitioning an image into a foreground and background. This image analysis technique is used to create binary images from a grayscale image. 

In this notebook, we have several images loaded as numpy arrays, such as `Seville`, `Seville_flipp`, `seville`, and `seville_fliph`. We can apply thresholding to these images to highlight specific features or regions of interest.

For example, we can apply a threshold to the `Seville` image to create a binary image where all pixel values above a certain threshold are set to one value (white), and all pixel values below that threshold are set to another value (black).

Let's proceed with applying a threshold to the `Seville` image.

Thresholding is used to partition the background and foreground of grayscale images, by essentially making them black and white. We compare each pixel to a given threshold value. If the pixel is less than that value, we turn it white. If it's greater; we turn it black

Let's see how to apply it. Once an image is loaded, we need to set the thresh value. Temporarily set it to 127, mid point between 0 and 255. We apply thresholding to an image by using the greater than operator followed by the thresh. Finally, show the thresholded image using show_image()





````python
#import scikit-image library
from skimage import color
import matplotlib.pyplot as plt

def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

#load the image
image = plt.imread('deer_iitm.jpg')
#convert the image to grayscale
image_gray = color.rgb2gray(image)
#show the image
show_image(image_gray, 'Grayscale image', 'gray')
#obtain the threshold image
thresh = 127.55555/255
binary = image_gray > thresh
#show the threshold image
show_image(binary, 'Binary image', 'gray')
````

# Comparing Plot

This is the plot comparing function we just used. It plots the original image next to the resulting one. Using this function allows us to reuse this code and focus more on the image processing part of the code.

The provided code defines a Python function named plot_comparison that is designed to display a side-by-side comparison of two images using the matplotlib library. This function is particularly useful for visualizing the differences between an original image and a processed version of that image, such as one that has undergone edge detection or other image processing techniques.

The function plot_comparison takes four parameters: original, processed, title_original, and title_processed. The original and processed parameters are the images to be compared. The title_original and title_processed parameters are optional strings that set the titles of the respective image plots, with default values of 'Original Image' and 'Processed Image'.

Inside the function, plt.subplots(1, 2, figsize=(12, 6)) is called to create a figure with two subplots arranged in a single row. The figsize parameter sets the size of the figure to 12 inches by 6 inches. The axes.ravel() method is then used to flatten the array of axes objects into a one-dimensional array, making it easier to index them.

The function then uses ax[0].imshow(original, cmap='gray') to display the original image in the first subplot with a grayscale colormap. The title of this subplot is set using ax[0].set_title(title_original), and the axis is turned on with ax[0].axis('on'). Similarly, the processed image is displayed in the second subplot using ax[1].imshow(processed, cmap='gray'), with its title set by ax[1].set_title(title_processed) and the axis turned on with ax[1].axis('on').

To ensure that the subplots are neatly arranged without overlapping, plt.tight_layout() is called. Finally, plt.show() is used to render and display the figure.

The function is then used to compare an original image and an edge-detected version of that image by calling plot_comparison(image, edges, title_original='Original Image', title_processed='Edges Detected'). This call displays the original image on the left and the edge-detected image on the right, allowing for an easy visual comparison.

````python
from skimage import filters, color

def plot_comparison(original, processed, title_original='Original Image', title_processed='Processed Image'):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    ax = axes.ravel()
    
    ax[0].imshow(original, cmap='gray')
    ax[0].set_title(title_original)
    ax[0].axis('on')
    
    ax[1].imshow(processed, cmap='gray')
    ax[1].set_title(title_processed)
    ax[1].axis('on')
    
    plt.tight_layout()
    plt.show()

# Ensure the variable 'image' is defined
image = plt.imread('image (1).png')
image1 = color.rgb2gray(image)
edges = filters.sobel(image1)

# Use the function to compare the original image and the edges detected image
plot_comparison(image, edges, title_original='Original Image', title_processed=' Edge detected with Sobel filter ')
````
