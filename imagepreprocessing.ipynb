Thresholding is a simple, yet effective, way of partitioning an image into a foreground and background. This image analysis technique is used to create binary images from a grayscale image. 

In this notebook, we have several images loaded as numpy arrays, such as `Seville`, `Seville_flipp`, `seville`, and `seville_fliph`. We can apply thresholding to these images to highlight specific features or regions of interest.

This code demonstrates a basic image processing workflow, including loading an image, converting it to grayscale, applying a threshold to create a binary image, and displaying the results at each step.

````python
#import scikit-image library
from skimage import color
import matplotlib.pyplot as plt

def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

#load the image
image = plt.imread('deer_iitm.jpg')
#convert the image to grayscale
image_gray = color.rgb2gray(image)
#show the image
show_image(image_gray, 'Grayscale image', 'gray')
#obtain the threshold image
#A threshold value is calculated by dividing 127.55555 by 255, which normalizes the threshold to the range [0, 1]. 
# This threshold is used to create a binary image, where pixels with intensity values greater 
# than the threshold are set to True (white), and others are set to False (black).
thresh = 127.55555/255
binary = image_gray > thresh
#show the threshold image
show_image(binary, 'Binary image', 'gray')
````

# Comparison Plot

This is the plot comparing function we just used. It plots the original image next to the resulting one. Using this function allows us to reuse this code and focus more on the image processing part of the code.

The provided code defines a Python function named plot_comparison that is designed to display a side-by-side comparison of two images using the matplotlib library. This function is particularly useful for visualizing the differences between an original image and a processed version of that image, such as one that has undergone edge detection or other image processing techniques.

The function plot_comparison takes four parameters: original, processed, title_original, and title_processed. The original and processed parameters are the images to be compared. The title_original and title_processed parameters are optional strings that set the titles of the respective image plots, with default values of 'Original Image' and 'Processed Image'.

Inside the function, plt.subplots(1, 2, figsize=(12, 6)) is called to create a figure with two subplots arranged in a single row. The figsize parameter sets the size of the figure to 12 inches by 6 inches. The axes.ravel() method is then used to flatten the array of axes objects into a one-dimensional array, making it easier to index them.

The function then uses ax[0].imshow(original, cmap='gray') to display the original image in the first subplot with a grayscale colormap. The title of this subplot is set using ax[0].set_title(title_original), and the axis is turned on with ax[0].axis('on'). Similarly, the processed image is displayed in the second subplot using ax[1].imshow(processed, cmap='gray'), with its title set by ax[1].set_title(title_processed) and the axis turned on with ax[1].axis('on').

To ensure that the subplots are neatly arranged without overlapping, plt.tight_layout() is called. Finally, plt.show() is used to render and display the figure.

The function is then used to compare an original image and an edge-detected version of that image by calling plot_comparison(image, edges, title_original='Original Image', title_processed='Edges Detected'). This call displays the original image on the left and the edge-detected image on the right, allowing for an easy visual comparison.

````python
from skimage import filters, color
import matplotlib.pyplot as plt

def plot_comparison(original, processed, title_original='Original Image', title_processed='Processed Image'):
    fig, axes = plt.subplots(1, 2, figsize=(5, 5)) ## size of the images in the comparison plot
    ax = axes.ravel() # ravel() returns a contiguous flattened array 
    
    ax[0].imshow(original, cmap='gray')
    ax[0].set_title(title_original)
    ax[0].axis('off')
    
    ax[1].imshow(processed, cmap='gray')
    ax[1].set_title(title_processed)
    ax[1].axis('off')
    
    plt.tight_layout() # automatically adjust subplot parameters to give specified padding
    plt.show() # display the images

# Ensure the variable 'image' is defined

image = plt.imread('image (1).png')
image1 = color.rgb2gray(image)
edges = filters.sobel(image1) # Apply the Sobel filter to the image to detect edges

#Sobel filter algorithm two CS SAIL Stanford Artificial Intelligence Lab researchers, Irwin Sobel and Gary Feldman, introduced the Sobel filter in 1968.
# The Sobel filter is used to detect edges in images.
# It works by calculating the gradient of the image intensity at each pixel. 
# The gradient is a vector that points in the direction of the greatest rate of change of intensity.
# The Sobel filter uses two 3x3 convolution kernels to calculate the gradient in the x and y directions.
# The x-direction kernel is used to detect vertical edges, while the y-direction kernel is used to detect horizontal edges.
# The Sobel filter is commonly used in computer vision and image processing applications to detect edges in images.

# Use the function to compare the original image and the edges detected image

plot_comparison(image, edges, title_original='Original Image', title_processed=' Edges detected with Sobel filter ')
````


## Gaussian Blur

Gaussian Blur is a widely used filter in image processing that smooths an image by reducing noise and detail. It works by convolving the image with a Gaussian function, which results in a weighted average of the surrounding pixels. This technique is particularly useful for reducing noise and preparing an image for further processing, such as edge detection.

### Applying Gaussian Blur

To apply Gaussian Blur to an image, we can use the `gaussian` function from the `skimage.filters` module. The function requires the image and the standard deviation (sigma) of the Gaussian kernel as input parameters.

Below is an example of how to apply Gaussian Blur to the `image1` variable, which is a grayscale version of the original image.

```python
from skimage.filters import gaussian

# Apply Gaussian Blur to the grayscale image
blurred_image = gaussian(image1, sigma=1)

# Show the original and blurred images
plot_comparison(image1, blurred_image, title_original='Original Grayscale Image', title_processed='Blurred Image with Gaussian Filter')
```

In this example, the `sigma` parameter is set to 1, which determines the amount of blurring. A higher sigma value results in more blurring. The `plot_comparison` function is used to display the original and blurred images side by side for easy comparison.
```
'''
Gaussian Blur and Gaussian Smoothing are interchangeable terms.
Both involve convolving an image with a Gaussian kernel to reduce noise and detail.
They are widely used in image processing for tasks like noise reduction, edge detection, and feature extraction.
'''
Lets see how we can do this with Scikit image using this picture of me in Amsterdam.

````python
#gaussian blur filter
from skimage.filters import gaussian
import matplotlib.pyplot as plt

def plot_comparison(original, processed, title_original='Original Image', title_processed='Processed Image'):
    fig, axes = plt.subplots(1, 2, figsize=(5, 3)) ## size of the images in the comparison plot
    ax = axes.ravel() # ravel() returns a contiguous flattened array 
    
    ax[0].imshow(original, cmap='gray')
    ax[0].set_title(title_original)
    ax[0].axis('off')
    
    ax[1].imshow(processed, cmap='gray')
    ax[1].set_title(title_processed)
    ax[1].axis('off')
    
    plt.tight_layout() # automatically adjust subplot parameters to give specified padding
    plt.show() # display the images

#load image
image = plt.imread('Screenshot 2024-10-03 225651.png')

#apply the gaussian filter
gaussian_image = gaussian(image, sigma=0.35) # sigma is the standard deviation of the Gaussian kernel and  
#controls the amount of smoothing applied to the image

#plot comparison
plot_comparison(image, gaussian_image, title_original='Original Image', title_processed='Gaussian Image')
````
